```python
def rentals_month (engine,year,month):
    month_str = f"{month:02d}"
    query = f'''SELECT *
    FROM sakila.rental
    WHERE SUBSTR(rental_date,1,7) = "{year}-{month_str}";'''
    
    data2 = pd.read_sql(query, engine)

    return data2
    
```

## Develop a Python function called rental_count_month that takes the DataFrame provided by rentals_month as input along with the month and year and returns a new DataFrame containing the number of rentals made by each customer_id during the selected month and year.

```python
def rental_count_month(df,month, year):
    
    df_filtered = df[(df['rental_date'].dt.month == month) & (df['rental_date'].dt.year == year)]
    
    result_df = df_filtered.groupby('customer_id').size().reset_index(name='rental_count')
    
    return result_df
    

    
```

## Create a Python function called compare_rentals that takes two DataFrames as input containing the number of rentals made by each customer in different months and years. 

```python
def compare_rentals(df1, df2):

    merged_df = df1.merge(df2, on='customer_id', how='outer', suffixes=('_date1', '_date2'))
    
    
    merged_df.fillna(0, inplace=True)
    
   
    merged_df['difference'] = merged_df['rental_count_date2'] - merged_df['rental_count_date1']
    
    return merged_df
```

```python
data_first = rentals_month(engine_sql,2006,2)
data_second = rentals_month(engine_sql,2005,5)
data_1 = rental_count_month(data_first,2,2006)
data_2 = rental_count_month(data_second,5,2005)
compare_rentals(data_1,data_2)
```
